# SVM examples

```python
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = [-1, 1, -1, -1, 1, 0, 2, 0, 3, 0]
X = np.array(a).reshape([-1, 2])
y = [1, 1, -1, -1, -1]
```

```python
def show_svm(kernel, train, test, gamma='auto', coef=1., deg=2):
    clf = svm.SVC(kernel=kernel, degree=deg, gamma=gamma, coef0=coef)
    clf.fit(train, test)
    xx, yy = np.meshgrid(np.arange(-5, 6, .01), np.arange(-3, 3, .01))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(train[:,0], train[:,1], c=test, cmap=plt.cm.coolwarm)
    plt.show()
    print '\n \n \n \n \n \n '
```

```python
kernels = ['linear', 'poly', 'rbf']

for k in kernels:
    show_svm(k, X, y)
```

```python
a_ = [-1, 1, -1, -1, 1, 0, 2, 0, 3, 0, 5, 0]
X_ = np.array(a_).reshape([-1, 2])
y_ = [1, 1, -1, -1, -1, 1]
```

```python
for k in kernels:
    show_svm(k, X_, y_)
```
